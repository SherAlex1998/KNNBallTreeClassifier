Настало время для реальной задачи. Будем классифицировать качество вина. Воспользуемся датасетом с осеннего курса Share.
Сравним качество работы самопального алгоритма и алгоритма с sklearn с теми же гиперпараметрами.

```python
trainDataFrame = pd.read_csv('train.csv')
testDataFrame = pd.read_csv('test.csv')

train = trainDataFrame.values
trainX = train[:, 0:11]
trainY = train[:, 11]

test = testDataFrame.values
testX = test[:, 0:11]
testY = test[:, 11]

print(trainX.shape)
print(trainY.shape)

print(testX.shape)
print(testY.shape)

model = KnnBalltreeClassifier()
model.fit(trainX, trainY)
predictClasses = model.predict(testX)
print("Вероятности самопального алгоритма: ")
print(model.predict_proba(testX))
trueClasses = testY
#print(predictClasses)
#print(trueClasses)
table = predictClasses == trueClasses
#print(table)
score = Counter(table)[True] / len(table)
print("accuracy самопального алгоритма: ")
print(score)
```

```python
baseline = KNeighborsClassifier(n_neighbors=1)
baseline.fit(trainX, trainY)
baseline_prediction = baseline.predict(testX)
print(baseline.predict_proba(testX))
print('accuracy = %.2f' % accuracy_score(testY, baseline_prediction))
```
